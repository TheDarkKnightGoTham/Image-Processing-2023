```python
import cv2 
import numpy as np
import matplotlib.pylab as plt
import math 

def showC(img):
    rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    plt.imshow(rgb,cmap="gray")
    
def show(img):
    plt.imshow(img,cmap="gray")
```

```python
from function import SelectShapeStd
```

```python
img = cv2.imread("images/lens.png")
gray = cv2.imread("images/lens.png",0)


_, threshold = cv2.threshold(gray,10, 255, cv2.THRESH_BINARY_INV)
threshold = SelectShapeStd(threshold, mode ="area")
show(threshold)
```

## Find Contours


* Mode Find:
1. Find list (all) 
2. Find external 
3. Hierarchical 
4. tree view 



* Mode point 
1. get all point
2. get first, second point on line 

```python
countours, _ = cv2.findContours(threshold, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

print(len(countours))
print(countours)
```

```python
img_copy  = img.copy() 
cv2.drawContours(img_copy, countours[0], -1, (0,0,255) ,2)
showC(img_copy)
```

## Lenght

```python
length =cv2.arcLength(countours[0], True)
print(length)
print(f'so canh ban dau : {len(countours[0])}')
```

## Approximate Contours

```python
epsilon = 3
approximate = cv2.approxPolyDP(countours[0], epsilon , True)

print(f'so canh cua contour: {len(approximate)}')
```

```python
approximate_contour =[]
approximate_contour.append(approximate)
img_copy = img.copy()
cv2.drawContours(img_copy, approximate_contour, -1, (0,0,255) ,2)
showC(img_copy)
```
