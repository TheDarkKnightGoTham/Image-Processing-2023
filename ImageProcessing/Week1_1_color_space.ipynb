```python
chanel_B  = img[:,:,0]
chanel_G  = img[:,:,1]
chanel_R  = img[:,:,2]
chanel_B.shape
```

```python
gray_value_B = chanel_B [100,200]
gray_value_G = chanel_G [100,200]
gray_value_R = chanel_R [100,200]

print(gray_value_B)
print(gray_value_G)
print(gray_value_R)

gray_value = int (0.1140 * gray_value_B + 0.5870 * gray_value_G + 0.2989 * gray_value_R)
print(gray_value)
```

# Create gray space and convert BGR to GRAY

```python
height = img.shape[0]
width = img.shape[1]
gray_space = np.zeros((height,width))
```

```python
for y in range(height):
    for x in range(width):
        gray_value_B = chanel_B [y,x]
        gray_value_G = chanel_G [y,x]
        gray_value_R = chanel_R [y,x]
        
        gray_value = int (0.1140 * gray_value_B + 0.5870 * gray_value_G + 0.2989 * gray_value_R)
        
        gray_space[y,x] = gray_value     
```

```python
gray_space.shape
```

```python
plt.imshow(gray_space,cmap="gray")
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray,cmap="gray")
```

# Decompose (Split channels)

```python
rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(rgb,cmap="gray")
```

```python
chanel_B  = img[:,:,0]
chanel_G  = img[:,:,1]
chanel_R  = img[:,:,2]
```

```python
plt.imshow(chanel_B,cmap="gray")
```

```python
plt.imshow(chanel_G,cmap="gray")
```

```python
plt.imshow(chanel_R,cmap="gray")
```

# Show Channel R

```python
R_space_show = np.zeros((img.shape[0],img.shape[1],3))

for y in range(img.shape[0]):
  for x in range(img.shape[1]):
    R_space_show[y,x][0] = 0
    R_space_show[y,x][1] = 0
    R_space_show[y,x][2] = chanel_R[y,x]

rgb = cv2.cvtColor(R_space_show.astype(np.uint8),cv2.COLOR_BGR2RGB)
plt.imshow(rgb,cmap ="gray")
```

# Show Channel G

```python
G_space_show = np.zeros((img.shape[0],img.shape[1],3))

for y in range(img.shape[0]):
  for x in range(img.shape[1]):
    G_space_show[y,x][0] = 0
    G_space_show[y,x][1] = chanel_G[y,x]
    G_space_show[y,x][2] = 0

rgb = cv2.cvtColor(G_space_show.astype(np.uint8),cv2.COLOR_BGR2RGB)
plt.imshow(rgb,cmap ="gray")
```

# Show Channel B

```python
B_space_show = np.zeros((img.shape[0],img.shape[1],3))

for y in range(img.shape[0]):
  for x in range(img.shape[1]):
    B_space_show[y,x][0] = chanel_B[y,x]
    B_space_show[y,x][1] = 0
    B_space_show[y,x][2] = 0

rgb = cv2.cvtColor(B_space_show.astype(np.uint8),cv2.COLOR_BGR2RGB)
plt.imshow(rgb,cmap ="gray")
```

# Compose 

```python
img_space_show = np.zeros((img.shape[0],img.shape[1],3))

for y in range(img.shape[0]):
  for x in range(img.shape[1]):
    img_space_show[y,x][0] = chanel_B[y,x]
    img_space_show[y,x][1] = chanel_G[y,x]
    img_space_show[y,x][2] = chanel_R[y,x]

rgb = cv2.cvtColor(img_space_show.astype(np.uint8),cv2.COLOR_BGR2RGB)
plt.imshow(rgb,cmap ="gray")
```

```python
image_show = cv2.merge((chanel_B,chanel_G,chanel_R))
rgb = cv2.cvtColor(image_show.astype(np.uint8),cv2.COLOR_BGR2RGB)
plt.imshow(rgb,cmap ="gray")
```
